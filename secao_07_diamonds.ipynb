```python
import seaborn as sns
import pandas as pd
```

```python
diamonds = pd.read_csv("Data Viz with Matplotlib & Seaborn/Data/Diamonds Prices2022.csv")
```

```python
diamonds.head()
```

```python
sns.barplot(
    x='carat', # se colocar a variavel numerica em X, automaticamente ser√° horizontal.
    y='cut',
    data=diamonds.sort_values(by='color'),
    palette='viridis',
    hue='color'
)
```

```python
sns.barplot(
    x='cut',
    y='carat',
    data=diamonds
)
```

```python
sns.histplot(
    x='price', 
    data=diamonds,
    bins=10,
    kde=True
    )
```

```python
sns.barplot(
    x='price',
    y='cut',
    data=diamonds,
    facecolor='white',
    edgecolor='black'
)
```

```python
sns.histplot(
    x='price',
    hue='clarity',
    data=diamonds.query("clarity in ['I1', 'IF']"),
    bins=10,
    binwidth=1000
)
```

```python
sns.histplot(
    x='price',
    hue='clarity',
    data=diamonds,
    cbar=True
)
```

```python
sns.scatterplot(
    x='x',
    y='price',
    data=diamonds,
    hue='carat',
    palette='husl',
    size='carat'
)
```
