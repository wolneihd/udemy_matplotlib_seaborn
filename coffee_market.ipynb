```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
coffee_production = pd.read_csv('Data Viz with Matplotlib & Seaborn/Data/coffee/total-production.csv', index_col='total_production')

# soma toda a linha, e salva em uma coluna nova.
coffee_production["Total"] = coffee_production.sum(axis=1)

top10_producers = coffee_production['Total'].sort_values(ascending=False).head(10)
```

```python
pd.DataFrame(top10_producers)
```

```python
# Iniciar
fig, ax = plt.subplots()

# configurar
ax.bar(
    top10_producers.index,
    top10_producers.values,
    width=.6
)

ax.set_title(
    'Top 10 coffee producers 1990-2018',
    weight='bold'
    )

ax.set_xlabel('Country')
ax.set_ylabel('Production (Millions 60kg bags)')

ax.xaxis.set_tick_params(rotation=30)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Gerar
plt.grid(ls=':')
plt.show()
```

```python
top_five = coffee_production.loc[top10_producers.index].drop(['Total'], axis=1).iloc[0:5]
top_five = top_five.transpose()
top_five.head()
```

```python
fig, ax = plt.subplots()

for i, column in enumerate(top_five.columns):
    ax.plot(
        top_five[column],
        label=top_five.columns[i],
    )

ax.set_title(
    'Top 5 coffee producers 1990-2018',
    weight='bold'
    )

ax.set_xlabel('Year')
ax.set_ylabel('Production (Millions 60kg bags)')

ax.set_xticks(top_five.index[::4])

ax.legend(loc='upper left')

plt.grid(ls=':')
plt.show()
```

```python
coffee_production = pd.read_csv('Data Viz with Matplotlib & Seaborn/Data/coffee/total-production.csv', index_col='total_production').transpose()
brazil = coffee_production['Brazil']
rest_world = coffee_production.drop(['Brazil'], axis=1).sum(axis=1)

pd.DataFrame({
    "brazil": brazil,
    'rest_world': rest_world
}).head()
```

```python
fig, ax = plt.subplots()

ax.stackplot(
    brazil.index,
    brazil,
    rest_world,
    labels=['Brazil', 'Rest World']
    )

fig.suptitle("Brazil's Share of Global Coffee Production 1990-2018")
ax.set_title("Brazil Increases share of Growing Market")
ax.set_ylabel("Production (60kg Bags)", fontsize=14)

ax.set_xticks(brazil.index[::4])

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.legend()
plt.show()
```

```python
coffe_raw = pd.read_csv('Data Viz with Matplotlib & Seaborn/Data/coffee/total-production.csv', index_col='total_production')
coffe_raw.head()
```

```python
coffee_production['total'] = coffe_raw.sum('columns')
rest_of_world = coffee_production.sort_values(by='total',ascending=False).iloc[5:]['total'].sum()
```

```python
top_procucers = coffee_production.sort_values(by='total',ascending=False)['total'].head()
```

```python
pie_data = (pd.concat([top_procucers, pd.DataFrame({"total": [rest_of_world]}, index=["Rest of World"])]))
```

```python
pie_data
```

```python
fig, ax = plt.subplots()

ax.pie(
    x=pie_data['total'],
    startangle=90,
    labels=pie_data.index,
    autopct="%.0f%%",
)

hole = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()

fig.gca().add_artist(hole)
ax.set_title('Share of Global Coffee Production 2018')
```
